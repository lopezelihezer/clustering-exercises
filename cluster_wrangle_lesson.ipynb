### Example:
#### Create a figure holding sin and cosine plots

```python
fig, (ax1, ax2) = plt.subplots(2,1)
# creates 1 figure containing two axes (two rows, one column)
# important: dont forget to create the figure first

# create a plot of y= sin(x) on the first row
x1 = np.linspace(0, 4 * np.pi, 100)
y1 = np.sin(x1)
ax1.plot(x1, y1)

# create a plot of y= cos(x) on the second row
x2 = np.linspace(0, 4*np.pi, 100)
y2 = np.cos(x2)
ax2.plot(x2, y2)

plt.show()
```

### Another way to do it!

```python
# Create the figure and two axes (two rows, one column)
fig, ax1 = plt.subplots(1, 1)

# Share the x-axis for both the axes (ax1, ax2)
ax2 = ax1.twinx()

# Create a plot of y = sin(x) on the first row
x1 = np.linspace(0, 4 * np.pi, 100) #returns 100 evenly spaced numbers between 0 and 4pi
y1 = np.sin(x1)
ax1.plot(x1, y1)

# Create a plot of y = cos(x) on the second row
x2 = np.linspace(0, 4 * np.pi, 100)
y2 = np.cos(x2)
ax2.plot(x2, y2)

# Save the figure
plt.show()
```

# Adding a legend!

```python
fig , ax1 = plt.subplots(1,1)

ax2 = ax1.twinx()

x1 = np.linspace(0, 4*np.pi, 100)
y1 = np.sin(x1)
# add a label for the legend
function1 = ax1.plot(x1, y1, label="Sine")

x2 = np.linspace(0, 4*np.pi, 100)
y2 = np.cos(x2)
function2 = ax2.plot(x2, y2, label="Cosine")

functions = function1 + function2
labels = [f.get_label() for f in functions]
plt.legend(functions, labels, loc=0)

plt.show()
```
